```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('valid_points.txt')
# m12, lp, lh, ls, l1, mH, mS1, mS2, PhysVev, theta}
# Columns indices 
M12_COL = 0
MS1_COL = 6
MS2_COL = 7
THETA_COL = 9

theta = data[:, THETA_COL]
ms1 = data[:, MS1_COL]
ms2 = data[:, MS2_COL]
m12 = data[:, M12_COL]
ms_min = np.minimum(ms1, ms2)

fig, axs = plt.subplots(2, 2, figsize=(12, 10), constrained_layout=True)

# 1. mS1 vs theta (theta on y-axis)
axs[0, 0].scatter(ms1, theta, s=10, alpha=0.7)
axs[0, 0].set_xlabel('$m_{S1}$', fontsize=14)
axs[0, 0].set_ylabel('$\\theta$', fontsize=14)
axs[0, 0].set_title('$m_{S1}$ vs $\\theta$', fontsize=16)
axs[0, 0].grid(True)

# 2. mS2 vs theta
axs[0, 1].scatter(ms2, theta, s=10, alpha=0.7)
axs[0, 1].set_xlabel('$m_{S2}$', fontsize=14)
axs[0, 1].set_ylabel('$\\theta$', fontsize=14)
axs[0, 1].set_title('$m_{S2}$ vs $\\theta$', fontsize=16)
axs[0, 1].grid(True)

# 3. min(mS1, mS2) vs theta
axs[1, 0].scatter(ms_min, theta, s=10, alpha=0.7)
axs[1, 0].set_xlabel('min($m_{S1}$, $m_{S2}$)', fontsize=14)
axs[1, 0].set_ylabel('$\\theta$', fontsize=14)
axs[1, 0].set_title('min(mS1, mS2) vs $\\theta$', fontsize=16)
axs[1, 0].grid(True)

# 4. -m12 vs theta
axs[1, 1].scatter(-m12, theta, s=10, alpha=0.7)
axs[1, 1].set_xlabel('$-m_1^2$', fontsize=14)
axs[1, 1].set_ylabel('$\\theta$', fontsize=14)
axs[1, 1].set_title('-$m_1^2$ vs $\\theta$', fontsize=16)
axs[1, 1].grid(True)

# Optional: remove unused subplot (bottom right)

plt.show()
```
